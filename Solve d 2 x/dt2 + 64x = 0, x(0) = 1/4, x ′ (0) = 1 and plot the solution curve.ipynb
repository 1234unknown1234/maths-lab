```python
import numpy as np
from scipy .integrate import odeint
import matplotlib .pyplot as plt
def f(u,x):
  return (u[ 1],- 64 *u[ 0])
y0=[ 1/4 ,1]
xs=np.linspace(0,5 ,50 )
us= odeint(f,y0 ,xs)
ys=us[:,0]
print(ys)
plt.plot(xs,ys,'r-')
plt.xlabel('Time ')
plt.ylabel('Amplitude ')
plt.title ('Solution of free and undamed case ')
plt.grid (True )
plt.show ()
```
